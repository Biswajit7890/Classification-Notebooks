```python
## Train & test Spli
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
Target='GoodCredit'
Predictors=['duration2','amount5','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_new[Predictors].values
y=creadit_risk_new[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=500)
```

```python
## logistic refression
from sklearn.linear_model import LogisticRegression
Target='GoodCredit'
Predictors=['duration2','amount5','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_new[Predictors].values
y=creadit_risk_new[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=4890)
lgf=LogisticRegression( C=4,penalty='l1', solver='liblinear')
predictmodel=lgf.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print(metrics.classification_report(y_test,predictions))
print(metrics.confusion_matrix(y_test,predictions))
print("Accuracy ", metrics.classification_report(y_test,predictions).split()[-2])



```

```python
## Decsion tree
from sklearn import tree
Target='GoodCredit'
Predictors=['duration2','amount5','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_new[Predictors].values
y=creadit_risk_new[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=5201)
dlf=tree.DecisionTreeClassifier(max_depth=3, criterion='gini')
predictmodel=dlf.fit(X_train,y_train)
prediction=predictmodel.predict(X_test) 
print(metrics.classification_report(y_test,prediction))
print(metrics.confusion_matrix(y_test,prediction))
print(metrics.classification_report(y_test,prediction).split()[-2]) 
feautre_importance=pd.Series(predictmodel.feature_importances_ , index=Predictors)
feautre_importance.nlargest(10).plot(kind='barh')   
```

```python
  ## Random forrest

from sklearn.ensemble import RandomForestClassifier
Target='GoodCredit'
Predictors=['duration2','amount5','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_new[Predictors].values
y=creadit_risk_new[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=5210)
rlf=RandomForestClassifier(n_estimators=300,criterion='gini',max_depth=8)
PredictModel=rlf.fit(X_train,y_train)
prediction =PredictModel.predict(X_test)
print(metrics.classification_report(y_test,prediction))
print(metrics.confusion_matrix(y_test,prediction))
print(metrics.classification_report(y_test,prediction).split()[-2]) 
feautre_importance=pd.Series(predictmodel.feature_importances_ , index=Predictors)
feautre_importance.nlargest(10).plot(kind='barh')      
```

```python
  ## Ada Boost
from sklearn.ensemble import AdaBoostClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
Target='GoodCredit'
Predictors=['amount5','duration2','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_Modify[Predictors].values
y=creadit_risk_Modify[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=5210)
dcf=DecisionTreeClassifier(max_depth=5)
ada=AdaBoostClassifier(n_estimators=300,base_estimator=dcf,learning_rate=0.2)
predictmodel=ada.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print(metrics.classification_report(y_test,predictions))
print(metrics.confusion_matrix(y_test,predictions))
print("Accuracy",metrics.classification_report(y_test,predictions).split()[-2])
feautre_importance=pd.Series(predictmodel.feature_importances_,index=Predictors)
feautre_importance.nlargest(12).plot(kind='bar')

```

```python
from IPython.display import Image
from sklearn import tree
import pydotplus

Dot_data =tree.export_graphviz(predictmodel.estimators_[10],out_file=None,feature_names=Predictors, class_names=Target)
Graph=pydotplus.graph_from_dot_data(Dot_data)

Image(Graph.create_png())

```

```python
##Xgboost
from xgboost import XGBClassifier
Target='GoodCredit'
Predictors=['amount5','duration2','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_Modify[Predictors].values
y=creadit_risk_Modify[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=510)
xgb=XGBClassifier(n_estimators=200,learning_rate=0.07,max_depth=6,objective='binary:logistic',booster='gbtree')
predictmodel=xgb.fit(X_train,y_train)
predictions=predictmodel.predict(X_test)
print(metrics.classification_report(y_test,predictions))
print(metrics.confusion_matrix(y_test,predictions))
print("Accuracy",metrics.classification_report(y_test,predictions).split()[-2])
feautre_importance=pd.Series(predictmodel.feature_importances_,index=Predictors)
feautre_importance.nlargest(12).plot(kind='bar')
```

```python
from xgboost import plot_tree
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(120,60))
plot_tree(predictmodel,num_trees=3, ax=ax)
```

```python
##KNN
from sklearn.neighbors import KNeighborsClassifier
Target='GoodCredit'
Predictors=['amount5','duration2','age13','checkingstatus1','purpose4','savings6','employ7','status9','property12','housing15','history3']
X=creadit_risk_Modify[Predictors].values
y=creadit_risk_Modify[Target].values
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=5100)
knf=KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)
predctmodel=knf.fit(X_train,y_train)
predictions=predctmodel.predict(X_test)
print(metrics.classification_report(y_test,predictions))
print(metrics.confusion_matrix(y_test,predictions))
print("Accuracy",metrics.classification_report(y_test,predictions).split()[-2])



```
