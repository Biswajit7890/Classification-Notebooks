```python
## Analyzing relation with Continuous varibale with Target categorical variable(ANNOVA TEST)
from scipy.stats import f_oneway
for i in cont_col:
    Annovacategory=Flight_data_New.groupby('ARR_DEL15')[i].apply(list)
    Annovaresult=f_oneway(*Annovacategory)
    print("The p value is",Annovaresult[1])
    if(Annovaresult[1]<=0.05):
         print('correleated')
    else:
        print('not correleated')
        


```

```python
## Analyzing relation with Categorical variable  with Target categorical variable(Chi square  TEST)

from scipy.stats import chi2_contingency
for i in cat_col:
    print('/'*60)
    crosstabresult=pd.crosstab(index=Flight_data_New[i],columns=Flight_data_New['DEP_DEL15'])
    print("The crosstabresult of ",i)
    print(crosstabresult)
    chiq_Result=chi2_contingency(crosstabresult)
    print("The P value is",chiq_Result[1])
    if(chiq_Result[1]<0.05):
        print(i,"correleated")
    else:
        print(i,"correleated")
        


```

```python
## interpolating continuous variable
Flight_data_New['DEP_TIME']=Flight_data_New['DEP_TIME'].interpolate(method='linear')
```

```python
Flight_data_New_ML=Flight_data_New[~((Flight_data_New<(Q1-1.5*IQR))|(Flight_data_New>(Q3+1.5*IQR))).any(axis=1)]
```

```python
Flight_data_New_ML=pd.DataFrame(Flight_data_New, columns=['DAY_OF_WEEK','OP_UNIQUE_CARRIER','DEST_AIRPORT_ID','DEP_TIME','DEP_DEL15','ARR_TIME','DIVERTED','DISTANCE','ARR_DEL15'])
```

```python
Flight_data_New_ML['DEST_AIRPORT_ID'].hist(figsize=(15,6))
```

```python
## Train and test split
from sklearn.model_selection import train_test_split
from sklearn import metrics

Predictors=['DAY_OF_WEEK','OP_UNIQUE_CARRIER','DEST_AIRPORT_ID','DEP_TIME','DEP_DEL15','ARR_TIME','DIVERTED','DISTANCE']
Target='ARR_DEL15'

X=Flight_data_New_ML[Predictors].values
y=Flight_data_New_ML[Target].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=5020)

```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression

X=Flight_data_New_ML[Predictors].values
y=Flight_data_New_ML[Target].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.31,random_state=5020)

lgf=LogisticRegression(C=1,penalty='l1', solver='liblinear')
predictModel=lgf.fit(X_train,y_train)
predictions=predictModel.predict(X_test)
print(metrics.classification_report(y_test, predictions))
print(metrics.confusion_matrix(y_test, predictions))
print('Accuracy of the model:', metrics.classification_report(y_test, predictions).split()[-2])






```

```python
#Decision Trees
from sklearn import tree


X=Flight_data_New_ML[Predictors].values
y=Flight_data_New_ML[Target].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.23,random_state=90)


dtree = tree.DecisionTreeClassifier(max_depth=2,criterion='entropy')
predictModel=dtree.fit(X_train,y_train)
predictions=predictModel.predict(X_test)
print(metrics.classification_report(y_test, predictions))
print(metrics.confusion_matrix(y_test, predictions))
print('Accuracy of the model:', metrics.classification_report(y_test, predictions).split()[-2])


feature_importances = pd.Series(dtree.feature_importances_, index=Predictors)
feature_importances.nlargest(10).plot(kind='barh')

```

```python
##Plotting a Decision Tree

from dtreeplt import dtreeplt
Ftree = dtreeplt(model=dtree, feature_names=Predictors, target_names=Target)
fig = Ftree.view()
```

```python
# Random Forest (Bagging of multiple Decision Trees)
from sklearn.ensemble import RandomForestClassifier


X=Flight_data_New_ML[Predictors].values
y=Flight_data_New_ML[Target].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.23,random_state=85470)

rgf = RandomForestClassifier(max_depth=3, n_estimators=150,criterion='entropy')
predictModel=rgf.fit(X_train,y_train)
predictions=predictModel.predict(X_test)
print(metrics.classification_report(y_test, predictions))
print(metrics.confusion_matrix(y_test, predictions))
print('Accuracy of the model:', metrics.classification_report(y_test, predictions).split()[-2])


feature_importances = pd.Series(rgf.feature_importances_, index=Predictors)
feature_importances.nlargest(10).plot(kind='barh')


```

```python
# PLotting 10th single Decision Tree from Random Forest
from dtreeplt import dtreeplt
Rtree = dtreeplt(model=rgf.estimators_[10], feature_names=Predictors, target_names=Target)
fig = Rtree.view()
```

```python
# Adaboost 
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier



X=Flight_data_New_ML[Predictors].values
y=Flight_data_New_ML[Target].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.40,random_state=724)


Dtree=DecisionTreeClassifier(max_depth=1)
ADA = AdaBoostClassifier(n_estimators=50, base_estimator=Dtree ,learning_rate=1)

PredictModel=ADA.fit(X_train,y_train)
predictions=predictModel.predict(X_test)
print(metrics.classification_report(y_test, predictions))
print(metrics.confusion_matrix(y_test, predictions))
print('Accuracy of the model:', metrics.classification_report(y_test, predictions).split()[-2])


feature_importances = pd.Series(ADA.feature_importances_, index=Predictors)
feature_importances.nlargest(10).plot(kind='barh')





```
